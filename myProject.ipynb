```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Load CSV file

try:
    df = pd.read_csv("scores.csv")
    print("[OK] scores.csv loaded successfully!")
except FileNotFoundError:
    print("[ERROR] scores.csv file not found.")
    raise SystemExit()

# 2. Display first 5 rows

print("\n[Dataset Preview]")
print(df.head())

# 3. Validate Score column

if "Score" not in df.columns:
    print("[ERROR] 'Score' column is missing in the dataset.")
    raise SystemExit()

# Score must be numeric
try:
    df["Score"] = pd.to_numeric(df["Score"])
except:
    print("[ERROR] Non-numeric values found in 'Score' column.")
    raise SystemExit()

# 4. Compute statistics

avg_score = df["Score"].mean()
max_score = df["Score"].max()
min_score = df["Score"].min()
median_score = df["Score"].median()

print("\n[Statistics]")
print("Average Score:", avg_score)
print("Maximum Score:", max_score)
print("Minimum Score:", min_score)
print("Median Score:", median_score)

# 5. Bar Chart

plt.figure()
plt.bar(df["Name"], df["Score"])
plt.xlabel("Name")
plt.ylabel("Score")
plt.title("Student Scores - Bar Chart")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("scores_barchart.png")
plt.show()
plt.close()
print("\n[OK] Bar chart saved as scores_barchart.png")

# 6. Pie Chart

plt.figure()
plt.pie(df["Score"], labels=df["Name"], autopct="%1.1f%%")
plt.title("Score Distribution - Pie Chart")
plt.axis("equal")
plt.savefig("scores_piechart.png")
plt.show()
plt.close()
print("[OK] Pie chart saved as scores_piechart.png")

# 7. Finish

print("\n[DONE] Program completed successfully!")
```
